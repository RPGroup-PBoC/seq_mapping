## Comparing two-point models with linear models

When using matrices to quantify transcription factor binding energies, it is common practice to use a linear model in which each sequence position contributes independently to the binding energy. This approach is used because of its simplicity, but it is known that it is inaccurate. Dinucleotide or two-point models, in which interactions between sequence positions are considered, are known to produce more accurate predictions. An advantage of our method is that we generate millions of mutants, most of which have two or more mutations within a binding site, which allows us to produce two-point matrix models in which one can consider all combinations of two mutations within the binding site. Here we assess whether this improves our predictive power. 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.gridspec as gridspec
import seaborn as sns
import sys
import pandas as pd
import random
import matplotlib.lines as mlines
import beeswarm as bs
from scipy.special import erf
from mpl_toolkits.axes_grid.inset_locator import inset_axes


# Import the project utils
sys.path.insert(0, '../utilities/')
import sortseq_utils as sortseq
import mwc_induction_utils as mwc
mwc.set_plotting_style()

# Set color scheme
colors=sns.color_palette('RdPu_r', 6)[1:]
colors_alt=sns.color_palette('Blues')
colors_qual=sns.color_palette(
    ['#bf0087', '#e5b600', '#40ac00', '#5233ea', '#bf3800', '#00acac'])

sns.set_palette(colors_qual)

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

We start by loading the two-point matrix in a readable format.

```python
# Make list of headers
two_point_head = pd.read_csv('../../data/models/O1_full_1027_operator_MCMC_PAIR_0_fixed', delimiter=' ', nrows=0)
headers = [name for name in two_point_head if (name.startswith('val')) | (name.startswith('pos'))]

# Load remainder of file
two_point_temp = pd.DataFrame(np.loadtxt('../../data/models/O1_full_1027_operator_MCMC_PAIR_0_fixed', skiprows=1))
two_point_temp.columns = headers
two_point_temp.head()
```

Next we tidy up the data to make it easier to work with.

```python
# First alter matrix so that values are in a single column 
pos_1 = []
pos_2 = []
for num in two_point_temp.pos:
    pos_1.append(int(np.floor(num)))
    pos_2.append(int(str('%0.2f' % num)[-2:]))
base = min(pos_1)
pos_1 = np.array(pos_1) - base
pos_2 = np.array(pos_2) - base

two_point = two_point_temp.drop('pos', axis=1)
two_point['pos_1'] = pos_1
two_point['pos_2'] = pos_2
two_point = pd.melt(two_point, id_vars = ['pos_1', 'pos_2'], value_vars = headers[1:])

# Take a look
two_point.head()
```

```python
# Separate first and second base into separate columns. 

base_1 = []
base_2 = []
for var in two_point.variable:
    base_1.append(var[-2])
    base_2.append(var[-1])
    
two_point = two_point.drop('variable', axis=1)
two_point['base_1'] = base_1
two_point['base_2'] = base_2

# Take a look
two_point.head()
```

```python
# Now clean things up a little more by reordering the columns
vals = two_point.value
two_point = two_point.drop('value', axis=1)
two_point['value'] = vals

# Take a look
two_point.head()
```

The dataframe is now set up so that we can use it to predict binding energies. The function that is used to make these predictions can be found in sortseq_utils.py.

```python
# Load measured operator data
measured_df = pd.read_csv('../../data/Onew_TF_titration_nonlinear_regression_summary.csv', index_col=0)

# Load sequence list
seqs = pd.read_csv('../../data/Onew_mutants_list.csv')

# replace "mut" with "O1new" in seqs list
for name in seqs.ID:
    seqs.ID.replace(name, 'O1new' + name[-3:], inplace=True)

# Add seqs column to measured_df
seq_list = []
for val in measured_df.operator:
    seq_list.append(seqs.seq[seqs.ID==val].values[0])
measured_df['seq'] = seq_list

# Add two point prediction column to measured_df
wt_en = -15.3
mult = 0.34184939991 # scaling factor for matrix
pred_list = []
for seq in measured_df.seq:
    pred_list.append(sortseq.two_point_binding_energy(seq, two_point, wt_en, mult))
measured_df['two_point_pred'] = pred_list

# Take a look
measured_df.head()
```

For the purposes of comparison, it will also be useful to add a column of predictions from a linear model that has been derived from the same data set. 

```python
# Import scaling factors for each energy matrix as a dataframe

mults = pd.read_csv('../../data/models/multipliers', delimiter='\t')

# Rename columns
mults.columns = ['operator', '1027', '1147', '446', 'RBS1']

# Melt dataframe into tidy data format
mults = pd.melt(mults, id_vars='operator', var_name='RBS', value_name='multiplier')

# Import relevant linear energy matrix and apply scaling factor
op = 'O1'
RBS = '1027'
multiplier = mults['multiplier'][(mults['operator']==op) & (mults['RBS']==RBS)].values[0]
linear_mat = multiplier * np.loadtxt('../../data/models/%s%s_fixed_linear_all' % (op, RBS), skiprows=1)

# Create list of linear model predictions and add to dataframe
wt_en = -15.3
linear_list = [sortseq.binding_energy(seq, linear_mat, wt_en) for seq in measured_df.seq.values]
measured_df['linear_pred'] = linear_list

# Take a look
measured_df.head()
```

```python
# Now let's plot the measured values against the predicted values and compare two-point and linear predictions 
fig, ax = plt.subplots(ncols=2, figsize=(8,4), sharey=True)
ax[0].plot([-16, -5], [-16, -5], ls='--', color='gray')
ax[1].plot([-16, -5], [-16, -5], ls='--', color='gray')
for name in measured_df.operator.unique():
    if name.startswith('O1new0'):
        ax[0].plot(measured_df.two_point_pred[measured_df.operator==name],\
                  measured_df.energy_kBT[(measured_df.operator==name)], 'o',\
                 markeredgewidth=2, markerfacecolor='white', markeredgecolor=colors_qual[0], zorder=9)
        ax[1].plot(measured_df.linear_pred[measured_df.operator==name],\
                  measured_df.energy_kBT[(measured_df.operator==name)], 'o',\
                 markeredgewidth=2, markerfacecolor='white', markeredgecolor=colors_qual[0], zorder=9)
    elif name.startswith('O1new1'):
        ax[0].plot(measured_df.two_point_pred[measured_df.operator==name],\
                  measured_df.energy_kBT[(measured_df.operator==name)], 's',\
                 markeredgewidth=2, markerfacecolor='white', markeredgecolor=colors_qual[1], zorder=9)
        ax[1].plot(measured_df.linear_pred[measured_df.operator==name],\
                  measured_df.energy_kBT[(measured_df.operator==name)], 's',\
                 markeredgewidth=2, markerfacecolor='white', markeredgecolor=colors_qual[1], zorder=9)
    elif name.startswith('O1new2'):
        ax[0].plot(measured_df.two_point_pred[measured_df.operator==name],\
                  measured_df.energy_kBT[(measured_df.operator==name)], '^',\
                 markeredgewidth=2, markerfacecolor='white', markeredgecolor=colors_qual[2], zorder=9)
        ax[1].plot(measured_df.linear_pred[measured_df.operator==name],\
                  measured_df.energy_kBT[(measured_df.operator==name)], '^',\
                 markeredgewidth=2, markerfacecolor='white', markeredgecolor=colors_qual[2], zorder=9)

ax[0].set_title('two-point model')
ax[1].set_title('linear model')
ax[0].set_xlabel(r'predicted $\Delta \varepsilon_R$ ($k_BT$)')
ax[1].set_xlabel(r'predicted $\Delta \varepsilon_R$ ($k_BT$)')
ax[0].set_ylabel(r'predicted $\Delta \varepsilon_R$ ($k_BT$)')

plt.tight_layout()
```

Based on a superficial analysis, the two-point model performs slightly better for 1 bp mutants, as it appears that the scaling factor is more accurate. However, the linear model appears to perform better for 2 bp and 3 bp mutants, which is surprising. Part of the explanation for this may be that our linear model is an "averaged" model in which the values for each position are averaged over many 2+ bp mutants, such that each position contains information derived from multiple-bp mutants. This stands in contrast to many linear models, in which the binding energies for single bp mutants are measured individually and the resulting energy matrix is used to predict binding energies for multiple bp mutants.
