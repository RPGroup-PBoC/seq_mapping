Here we will look at the output of parallel tempering chains which try to maximize the mutual info between expression and bin number.

There are a couple main issues we need to address here. They are
     1. Has the chain converged?
     2. What thinning factor do we need to acheive independent measurements?
     3. Is the answer we get out physically reasonable (and if we have a point of comparison is it accurate)?
     
Previously I have used the definition of convergence that if you run multiple chains starting from different starting values and (after a burn in period) the posteriors they produce are identical the chains have converged. However, this definition only works if your starting points are widely distributed compared to the posterior. My starting points were not. That combined with occasionally extermely long autocorrolation times (which I didn't properly account for) made it appear as if the run had converged when it had not truly done so. 

Let's first look at autocorrolation times as according the the creators of emcee this is the most important statistic when looking at your runs. Theoretically your runs should converge after about 2-3 autocorrolation times, and if you sample for a few more you should have enough independent measurements to constrain the parameters you are inferring.

```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import glob
import ecdf
import pandas as pd
from scipy.optimize import minimize


#the next five functions are from a post from the creators of emcee. They will calculate the autocorrolation time
#for a chain.
def next_pow_two(n):
    i = 1
    while i < n:
        i = i << 1
    return i

def autocorr_func_1d(x, norm=True):
    x = np.atleast_1d(x)
    if len(x.shape) != 1:
        raise ValueError("invalid dimensions for 1D autocorrelation function")
    n = next_pow_two(len(x))

    # Compute the FFT and then (from that) the auto-correlation function
    f = np.fft.fft(x - np.mean(x), n=2*n)
    acf = np.fft.ifft(f * np.conjugate(f))[:len(x)].real
    acf /= 4*n

    # Optionally normalize
    if norm:
        acf /= acf[0]

    return acf

# Automated windowing procedure following Sokal (1989)
def auto_window(taus, c):
    m = np.arange(len(taus)) < c * taus
    if np.any(m):
        return np.argmin(m)
    return len(taus) - 1

# Following the suggestion from Goodman & Weare (2010)
def autocorr_gw2010(y, c=5.0):
    f = autocorr_func_1d(np.mean(y, axis=0))
    taus = 2.0*np.cumsum(f)-1.0
    window = auto_window(taus, c)
    return taus[window]

def autocorr_new(y, c=5.0):
    f = np.zeros(y.shape[1])
    for yy in y:
        f += autocorr_func_1d(yy)
    f /= len(y)
    taus = 2.0*np.cumsum(f)-1.0
    window = auto_window(taus, c)
    return taus[window]

```

These autocorrolation estimators work very well when the chain length is 50 times the autocorrolation time or greater. We run for 10000 iterations and the typical autocorrolation time as we will see below is 200-400 (if you take into account that the chains are already thinned by a factor of 10). There are also some with much longer predicted autocorrolation times. In the future we should see if we can find another estimator, or run these chains longer/with better settings to make sure we are getting the proper autocorrolation values.

```python
#first load in files in h5py format, and remove incorrect files from names list.
names = glob.glob('*')
names = [x for x in names if '.ipynb' not in x and 'acceptance' not in x and 'chain' in x]
autotimes = []
burnin = 200 
print(names)
for name in names:
    print(name)
    df = h5py.File(name,'r')
    em = df['chain']
    '''I have tested multiple temperature ladders to see which acheive better mixing. We only want to measure
         using the lowest temperature ladder though. Here we will define which one that is.'''
    if em.shape[0] == 10:
        prop_temp = 9
    else:
        prop_temp = 0
    all_IM = em[prop_temp,:,100:,3].ravel()
    all_nz = all_IM > 100
    all_IM = all_IM[all_nz.nonzero()]
    b = plt.hist(all_IM)
    plt.show()
    
    #read in the trace of each walker, we will use feature 1 for calculating autocorrolation.
    trace = em[prop_temp,:,burnin:,1]
    #we are running this on some chains which haven't reached the full 10000 iterations yet.
    #this shouldn't affect the calculation, but we do need to remove trailing zeros.
    '''
    chain_iterations = len(trace[0,:].nonzero()[0])
    traces = np.zeros((trace.shape[0],chain_iterations))
    for i in range(trace.shape[0]):
        traces[i,:] = trace[i,:][trace[i,:].nonzero()[0]]
        plt.plot(range(len(traces[i])),traces[i])
    '''
    traces = []
    for i in range(trace.shape[0]):
        traces.append(trace[i,:][trace[i,:].nonzero()[0]])
        plt.plot(range(len(traces[i])),traces[i])
    traces = np.array(traces)
    
    plt.ylabel('Feature 1 value')
    plt.xlabel('Iteration (thinned by a factor of 10)')
    plt.show()

    #now find autocorrolation time using both methods supplied by the creators of emcee. The 'new' method
    #should be better.
    
    #some runs will have failed or will not be far enough along to calculate autocorrolation time.
    try:
        gwk = autocorr_gw2010(traces)
    except:
        gwk = np.inf
    
    try:
        new = autocorr_new(traces)
    except:
        new = np.inf
    print('gwk autotime = ' + str(gwk))
    print('new autotime = ' + str(new))
    autotimes.append(new)
```

We can look at two ways to measure convergence from this. The Geweke score is one measure of convergence. It compares the beginning of the chain to the end section of a chain. If the chain has converged the calculated values should fall on a normal curve. The Geweke score essentially tells you how many stardard deviations away from the mean the subchain is. Lower scores are better and anything with a magnitude higher than 2 is really bad. All the chains look good on geweke score except O1_chain_10 and O1_chain_11. These two chains also have a very long autocorrolation time. I would use O1_1027_chain15 for calculating values for now.

The second measure of convergence is whether or not chains starting from different points are now identical. O1_1027 chains 10 and 11 once again do a bad job of this, while some others (O1_1147, O2_446) do quite well. 

It seems as if the new temperature ladder works better than the old for achieving good mixing amongst different temperature chains and getting a low autocorrolation time.
